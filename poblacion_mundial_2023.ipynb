# Renombrar columnas

```python
df.rename(columns={"Country (or dependency)": "Country_(OrDependency)", "Population\n(2023)": "Population_2023", "Yearly\nChange": "Yearly_Change(%)", "Net\nChange": "Net_Change", \
                   "Density\n(P/Km²)": "Density_(P/Km²)", "Land Area\n(Km²)": "LandArea_(Km²)", "Migrants\n(net)": "Migrants_(net)", "Fert.\nRate": "Fert_Rate", \
                   "Med.\nAge": "Med_Age", "Urban\nPop %": "Urban_Pop(%)", "World\nShare": "World_Share(%)"}, inplace=True)
df.drop(["#"], axis=1, inplace=True)
df.head(10)
```

```python
df.info()
```

# Filtrado de filas con valores faltantes en tasas de fertilidad y edad media

```python
FertNan_filter = df["Fert_Rate"].isna()
AgeNan_filter = df["Med_Age"].isna()
df.loc[FertNan_filter | AgeNan_filter, :]
```

# Actualización de datos para el Estado de la Ciudad del Vaticano (Santa Sede)

```python
df1=df
HolySee_filter = df1["Country_(OrDependency)"] == "Holy See"
df1.loc[HolySee_filter, "Fert_Rate"] = 1.0
df1.loc[HolySee_filter, "Med_Age"] = 57
df1.loc[HolySee_filter, :]
```

```python
df1.info()
```

# Eliminación de comas en la población mundial para el año 2023 y Eliminación de porcentajes en el cambio anual

```python
def delete_comma(column):
  try:
    value = column.replace(",", "")
    return int(value)
  except:
    return value

df1["Population_2023"] = df1.apply(lambda row: delete_comma(row["Population_2023"]), axis=1)
```

```python
def delete_percentage(column):
  try:
    value = column.replace("%", "")
    value = value.strip("% \t\n\r\x0b\x0c")
    return float(value)
  except:
    return value

df1["Yearly_Change(%)"] = df1.apply(lambda row: delete_percentage(row["Yearly_Change(%)"]), axis=1)
```

```python
df1["Density_(P/Km²)"] = df1.apply(lambda row: delete_comma(row["Density_(P/Km²)"]), axis=1)
df1["Urban_Pop(%)"] = df1.apply(lambda row: delete_percentage(row["Urban_Pop(%)"]), axis=1)
df1["World_Share(%)"] = df1.apply(lambda row: delete_percentage(row["World_Share(%)"]), axis=1)
df1.info()
```

# Filtrado de filas con valores "N.A." en la población urbana

```python
na_filter = df1["Urban_Pop(%)"] == "N.A."
df1.loc[na_filter, :]
```

# Preparación de datos para visualización

```python
df2 = df1.loc[0:19, ["Country_(OrDependency)", "Population_2023"]]
df2["Population_2023"] = ((df2["Population_2023"]/1000000).round()).astype(int)
countries = np.array(df2["Country_(OrDependency)"])
population_mill = np.array(df2["Population_2023"])
```

# Gráfico de donut de las 20 poblaciones más grandes del mundo

```python
for i in range(len(countries)):
  countries[i] = f"{countries[i]} {population_mill[i]}M"
fig, ax = plt.subplots(figsize=(10, 10));
ax.pie(population_mill, labels=countries, startangle=120, counterclock=False);
countries_legend = np.array(df2["Country_(OrDependency)"]).tolist()
ax.set_title("Top 20 Population in the World", fontsize=15)
ax.legend(labels = countries_legend, loc="center left", bbox_to_anchor=(-0.3, 0.5), title="Country Name")
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)
plt.show()
```
